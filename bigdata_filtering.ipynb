```python
words = r.split()
rst = [word for word in words if word.lower() not in bad_words]
rst = ' '.join(rst)

wordcloud = WordCloud(max_font_size=40).generate(rst)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
